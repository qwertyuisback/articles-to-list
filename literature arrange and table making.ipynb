```python
import re

# Read the original file
with open("lit.txt", "r", encoding="utf-8") as f:
    text = f.read()

# Remove "Save" sections (from 'Save' to 3 consecutive newlines)
text = re.sub(
    r'Save\n\[https:\/\/scholar\.google\.com\/intl\/en\/scholar\/images\/1x\/save-32\.png\]https:\/\/scholar\.google\.com\/citations\?.*?\n\n\n',
    '',
    text,
    flags=re.S
)

# Remove Google Scholar alert sections
text = re.sub(
    r"This message was sent by Google Scholar because you're following new results for\n\[.*?\]\"",
    '',
    text,
    flags=re.S
)

with open("articles_cleaned_output_emailend.txt", "w", encoding="utf-8") as f:
    f.write(text)

print("✅ Cleaned file saved as 'articles_cleaned_output_emailend.txt'")
```

```python
#merge article names
from datetime import datetime
import re
filetime=datetime.now().strftime("%Y%m%d-%H%M%S")

def is_uppercase_line(line):
    """
    Return True if the line is all uppercase (ignoring punctuation and numbers).
    """
    text_only = re.sub(r'[^A-Za-z]+', '', line)
    return text_only.isupper() and len(text_only) > 0

with open("articles_cleaned_output_emailend.txt", "r", encoding="utf-8") as f:
    content = f.read()

blocks = re.split(r'\n{2,}', content.strip())

merged_blocks = []
for block in blocks:
    lines = [l.strip() for l in block.split('\n') if l.strip()]
    if not lines:
        continue

    merged_title_lines = []
    remaining_lines = []
    found_body = False

    for line in lines:
        if not found_body and is_uppercase_line(line):
            merged_title_lines.append(line)
        else:
            found_body = True
            remaining_lines.append(line)

    merged_title = " ".join(merged_title_lines).strip()
    merged_text = "\n".join([merged_title] + remaining_lines)
    merged_blocks.append(merged_text.strip())

final_text = "\n\n".join(merged_blocks)
filename="articles_namemerged"+"-"+filetime+".txt"
# Save output
with open(filename , "w", encoding="utf-8") as f:
    f.write(final_text)

print("✅ Titles merged successfully and saved as:","articles_namemerged"+"-"+filetime+".txt")
```

```python
import re
import pandas as pd
with open(filename, "r", encoding="utf-8") as f:
    text = f.read()

articles_raw = re.split(r"\n{2,}", text.strip())
print(articles_raw, "\n")

article_table=pd.DataFrame(columns=['article','link','author','abstract'])
a=0
for i in range(0, len(articles_raw), 2): #Process each article, 2 items at a time
    titleandlink = articles_raw[i].strip()
    # print("Titleandlink:", titleandlink)
    parts = titleandlink.split("\n", 1)
    article_name = parts[0].strip()
    link = parts[1].strip() if len(parts) > 1 else ""
    # print("Article name:", article_name)
    # print("Link:", link)

    # author + abstract
    if i + 1 < len(articles_raw):
        authorandabstract = articles_raw[i + 1].strip()
        author_lines = authorandabstract.split("\n")
        author = author_lines[0].strip()
        abstract_lines = author_lines[1:]
        abst_text = " ".join([a.strip() for a in abstract_lines])

        # print("author:", author)
        # print("abstract:", abst_text)
    else:
        author = ""
        abst_text = ""

    # Print the dictionary
    article_dict = {
        "article": article_name,
        "link": link,
        "author": author,
        "abstract": abst_text
    }
    # print(article_dict)
    article_table.loc[a]=[article_name,link,author,abst_text]
    a+=1

filetime=datetime.now().strftime("%Y%m%d-%H%M%S")

article_table.to_excel("article_table-"+filetime+".xlsx")
article_table.head(25)
```
